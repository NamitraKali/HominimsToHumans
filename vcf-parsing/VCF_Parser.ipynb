```python
import pandas as pd
```

```python
# read the VCF using Pandas
df = pd.read_csv('All_merged.csv', sep='\t')

# Display the table
df.iloc[0:20]
```

```python
# Display last 20 rows
df.iloc[-20:]
```

```python
# Get all the names of the individuals in this table
columns = list(df)[9:]
```

```python
# Write the haplotypes as a fasta file
muts = [[df['REF'].iloc[i]] + df['ALT'].iloc[i].split(',') for i in range(len(df))]

aln_file = open('1000denisovans.fasta', 'w')

for header in columns:
    variants = df[header]
    seq = ''
    for i in range(len(muts)):
        index = variants.iloc[i]
        seq += muts[i][variants.iloc[i]]
 
    aln_file.write(f'>{header}\n{seq}\n')

aln_file.close()
```
