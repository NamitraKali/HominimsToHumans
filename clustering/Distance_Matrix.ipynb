```r
library(reshape)
library(seqinr)
library(ape)
```

```r
myseqs <- seqinr::read.alignment('1000denisovans_aln.fasta', 'fasta')
aln <- ape::as.alignment(myseqs)
bin <- ape::as.DNAbin(myseqs)
```

```r
aln2 <- seqinr::dist.alignment(myseqs, matrix = "identity" )
mat2 <-  as.matrix(aln2)
mat2
```

```r
write.csv(mat2, file='identityDist.csv')
```
